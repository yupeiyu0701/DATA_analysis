上述函数中常用参数的含义如下：<br>
(1)a:表示要观察的数据，可以是Series，一堆数组或列表<br>
(2)bins:用于控制条形的数量<br>
（3）hist：接收布尔类型，表示是否绘制（标注）直方图


```python
np.random.seed(80)
arr=np.random.randn(100)
sns.histplot(arr,bins=10,kde=True)
```
